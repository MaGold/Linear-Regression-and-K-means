#Normal Distribtuion

##$N(10, 2)$

```python
mu = 10
sigma = 2
```

```python
n = 2*np.random.randn(1000000)+mu
```

```python
p = plt.hist(n, bins=100)
```

##Standard normal

```python
z = (n-mu)/sigma
p = plt.hist(z, bins=100)
```

##$\chi^2$-distribution

$z_1^2 + z_2^2 + \cdots z_{\nu}^2$ is $\chi^2(\nu)$-distributed with $\nu$ degrees of freedom

```python
nu = 3
chi = []
for i in range(len(z)):
    chi.append(np.sum(z[i:i+nu]**2))
    i = i+nu
```

```python
p = plt.hist(chi, bins=100)
```

##$t$-distribution:
If $z$ is stadnard normal, and $\chi^2$ is $\chi^2$ with $\nu$ degrees of freedom, and they are independent, then $$t=\frac{z}{\left[\chi^2(\nu) / \nu\right]^{1/2}}$$ follows the $t$-distribution.

```python
t = []
for i in range(len(chi)):
    t.append( np.random.randn() / np.sqrt(chi[i]/nu))
```

```python
p = plt.hist(t, bins=np.linspace(-5, 5, 100))
```

##$F$-Distribution:
If $X_1 \sim \chi^2(\nu_1)$, and $X_2 \sim \chi^2(\nu_2)$ are independent, then
$$f = \frac{X_1}{\nu_1} / \frac{X_2}{\nu_2}$$
follows the $F$-distribtuion (with parameters ($\nu_1, \nu^2$).

```python
chi1 = []
chi2 = []
nu1 = 5
nu2 = 7
F=[]
for i in range(100000):
    chi1 = np.sum(np.random.randn(nu1)**2)
    chi2 = np.sum(np.random.randn(nu2)**2)
    F.append( (chi1/nu1) / (chi2/nu2) )
    
```

```python
p = plt.hist(F, bins=np.linspace(-10, 10, 100))
```
